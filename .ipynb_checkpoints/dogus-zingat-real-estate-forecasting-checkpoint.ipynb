# Data preprocessing,Visualization and Feature Engineering

```python
import pandas_datareader.data as web
# web.DataReader ile Dolar-TRY kur değerlerini çekiyoruz
curr=web.DataReader(['USDTRY=X'],'yahoo')['Adj Close'].rename(columns={'USDTRY=X':'USD'}).reset_index() 
dum=pd.date_range("2016-10-23",'2021-10-21', freq="d")
dum=pd.DataFrame(dum,columns=['Date'])

curr=dum.merge(curr,on='Date',how='left').ffill().dropna()
curr=curr.set_index(curr.Date)
curr.drop('Date',axis=1,inplace=True)
dolar=curr.reset_index().rename(columns={'Date':'tarih'})

train=pd.read_csv('zingat_usecase_trainset.csv').rename(columns={'brütm²':'brütm', 'netm²':'netm'})
test=pd.read_csv('zingat_usecase_testset_null.csv').rename(columns={'brütm²':'brütm','netm²':'netm'})

train=train[test.columns]
data=pd.concat([train,test],axis=0)
data.replace('-',np.nan,inplace=True)
data.replace('_',np.nan,inplace=True)
data.reset_index(drop=True,inplace=True)
data.odasayısı=data.odasayısı.astype('O')
data.güncel_fiyat=data.güncel_fiyat.map(lambda x :  int(x.split()[0]) if pd.notna(x)  else x)
data.tarih=pd.to_datetime(data.tarih)

# path
#bazı path değerlerinde ilce ve semt değerleri yok bunlara nan atarak 3 tana il-ilce-sem değişkni oluşturuyorum
data['il']=data.path.map(lambda x : x.split('/')[0] )
data['ilce']=data.path.map(lambda x : x.split('/')[1]  if len(x.split('/'))>1 else np.nan)
data['semt']=data.path.map(lambda x : x.split('/')[2]  if len(x.split('/'))>2 else np.nan)
data.drop(['path'],axis=1,inplace=True)

# data=pd.concat([data[data.güncel_fiyat.notnull()].drop_duplicates(subset=data.columns.difference(['tarih','Ilan_ID']),keep='first'),data[data.güncel_fiyat.isnull()]])  
# data.drop(data[(data.binayaşı.isna() ) & (data.bulunduğukat.isna()) & (data.binadakikatsayısı.isna())&(data.isıtmatipi.isna()) & (data.güncel_fiyat.notnull())].index,inplace=True)
# data.drop(data[(data.güncel_fiyat.notnull()) & (data.ilce.isnull())].index,inplace=True)
data=data.merge(dolar,how='left',on='tarih').set_index(data.index) #dolar kuru ile datamızı merge ediyoruz.
data.drop(['otopark','manzara'],axis=1,inplace=True)
```

# netm ve brütm 

```python
data.netm=data.apply(lambda x :  x.brütm if x.netm>x.brütm  else x.netm,axis=1) #netm eğer brütm den büyükse brütm yi netm ye eşitliyorum
data.loc[data.odasayısı=='1149+0','odasayısı']=np.nan # odasayısı değişkninde bulunan '1149+0' değerine nan atıyorum
data.odasayısı=data.odasayısı.map(lambda x:  sum(  int(i) for i in x.split('+')) if pd.notna(x) else x) #odasayısı değişkenimizi object değerden int değere çeviriyoruz örneğin '2+1'=3 gibi
 
data.netm=data.apply(lambda x:   x.brütm  if (x.netm*3<x.brütm) & (x.netm<25) else x.netm ,axis=1) #brütm değeri netm değerinden 3 kat fazla ve netm değeri 25 den küçük olan netm değerlerini brütm değerine eşitliyorum
data.loc[data.brütm>data.netm*4,'brütm']=data.netm*2 #brütm değğeri netm değeriin 4 katından büyük brütm değerlerini netm*2 ye eşitliyorum

data.loc[(data.netm/data.odasayısı)>80,'netm']=data.loc[(data.netm/data.odasayısı)>80].odasayısı*35
data.loc[(data.brütm/data.odasayısı)>150,'brütm']=data.loc[(data.brütm/data.odasayısı)>150].netm*2
data.loc[data.netm<12,'netm']=data.loc[data.netm<12].odasayısı*35
data.loc[data.brütm<12,'brütm']=data.loc[data.brütm<12].netm*2

data.drop(data[data.güncel_fiyat.notnull()&data.odasayısı.isnull()].index,axis=0,inplace=True)
data.odasayısı=data.odasayısı.map(lambda x: 10 if x>10 else x)   #odasayısı 10  dan fazla olan değerleri 10 a eşitiliyorum
data.loc[data.odasayısı.isnull(),'odasayısı']=data[data.odasayısı.isnull()].netm//35
data.loc[data.odasayısı>10,'netm']=10*35
data.odasayısı=data.odasayısı.map(lambda x: 10 if x>10 else x)
data.odasayısı=data.odasayısı.map(lambda x: 1 if x==0 else x)
data.loc[(data.emlaktipi!='Daire')&(data.emlaktipi!='Prefabrik Ev'),'brütm']=data.loc[(data.emlaktipi!='Daire')&(data.emlaktipi!='Prefabrik Ev'),'netm']*2

plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.boxplot(x=data.netm)
plt.subplot(1,2,2)
sns.boxplot(x=data.brütm)
set({'figure.figsize':(10,8)});
```

```python
dum=data[(data.güncel_fiyat.notnull()) &(data.güncel_fiyat>80000)].groupby(['il','ilce','emlaktipi']).agg({'brütm':'sum','güncel_fiyat':'sum'})
dum['brütm_fiyat']=dum.güncel_fiyat/dum.brütm
data.güncel_fiyat=data.apply(lambda x :  dum.loc[x.il,x.ilce,x.emlaktipi].brütm_fiyat*x.brütm if (x.güncel_fiyat<100000) & (x.emlaktipi !='Prefabrik Ev')&((x.il,x.ilce,x.emlaktipi) in dum.index)  else x.güncel_fiyat, axis=1)
data.drop(data[ (data.güncel_fiyat<100000) & (data.emlaktipi !='Prefabrik Ev')].index,axis=0,inplace=True)


#bazı ilçelerde ortalamdan çok yüksekte olan  emlak değerlerini düşürdüm
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Bağcılar')].index,axis=0,inplace=True)   
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Beylikdüzü')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Pendik')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Balçova')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Bornova')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Torbalı')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Seferihisar')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Karşıyaka')].index,axis=0,inplace=True)
data.drop(data[(data.güncel_fiyat>10000000) & (data.ilce=='Güzelbahçe')].index,axis=0,inplace=True)

data.drop(data.where( (data.güncel_fiyat>250000000.0)).dropna(how='all').index,inplace=True)
sns.set(rc = {'figure.figsize':(10,5)})
sns.boxplot(x='güncel_fiyat',y='emlaktipi',data=data); 
```

```python
# NAN olan ilce ve semt değerlerine atamalar yapıldı burda en sık olan ilce ve semtler dikkate alındı
data.loc[data.ilce.isnull(),'ilce']='Karşıyaka'
data.loc[data.semt.isnull(),'semt']='Bostanlı'
data.loc[data.ilce.isin(['Beydağ','Kınık','Kiraz']),'ilce']='Karşıyaka'
data.loc[data.ilce.isin(['Beydağ','Kınık','Kiraz']),'semt']='Bostanlı'
```

```python
# il_ilce_semt olarak yeni bir 'konum' değişkeni oluşturuldu
data['konum']=data.apply(lambda x:  x.il+'_'+x.ilce+'_'+x.semt,axis=1 )
```

```python
# elimizde bulunan güncel_fiyat değerlerini  'il','ilce','semt','emlaktipi' göre gruplayıp  dolar kuru cinsinden o günkü m2 fiyatı hesaplandı daha sonra hesaplanan bu m2 değerlei brütm m2 ile çarpılarak bütün ilanlar için yeni bir 'usd_fiyat' değişkeni oluşturuldu  bu değişkenimiz fiyat ile korelasyonu yüksek bir değişken oldu
data['fiyat_usd']=data.güncel_fiyat/data.USD
dum=data[(data.güncel_fiyat.notnull())].groupby(['il','ilce','semt','emlaktipi']).agg({'brütm':'sum','fiyat_usd':'sum'})
dum['brütm_fiyat_usd']=dum.fiyat_usd/dum.brütm
data['usd_fiyat']=data.apply(lambda x: dum.loc[x.il,x.ilce,x.semt,x.emlaktipi].brütm_fiyat_usd*x.brütm *x.USD  if ((x.il,x.ilce,x.semt,x.emlaktipi) in dum.index) else np.nan,axis=1)
data['usd_fiyat']=data['usd_fiyat'].fillna(data.usd_fiyat.mode()[0])
data.drop('fiyat_usd',axis=1,inplace=True)
#tarih değişkeninde günleri 1 yaptım aslında burda yıl ve ay'a göre gruplama yapmış gibi oldum 2019-01 2019-02 2020-01 gibi her ayı ayrı temsil etmiş olduk
data['yıl_ay']=data.tarih.map(lambda x : x.replace(day=1))
data['yıl_ay']=data['yıl_ay'].astype('O') 
```

```python
# güncel_fiyat değişkenine göre konumlarımızı grupluyorum fiyat olarak birbirine yakın olan konumlar aynı gruplara alınıyor.
from sklearn.cluster import KMeans
dff=data[data.güncel_fiyat.notnull()].groupby('konum')['güncel_fiyat'].mean().reset_index()
dff.set_index(dff.konum,inplace=True)
dff.drop('konum',axis=1,inplace=True)
Kmean = KMeans(n_clusters=40)
Kmean.fit(dff)
dff['semt_label']=Kmean.labels_
dff=dff.reset_index()
dff=dff[['konum','semt_label']]
dff=dff.drop_duplicates()

#Bu noktadan sonra data dataframe değişmesin diye yeni bir data2 frame i oluşturup devam ediyorum.
data2=data.merge(dff,how='left',on='konum')
data2.semt_label=data2.semt_label.map(lambda x :  'bilinmiyor' if pd.isna(x) else x)

# -------------------------------------------------------------------------------------------------------------------------------------------------------------
# gelişmişlik skorlarına göre yine konumları grupluyorum
gelişmişlik=pd.read_excel('/kaggle/input/gelimilik2/gelimilik_dzey.xlsx')#dış veri
#birtakım önişleme işlemleri 
gelişmişlik.ilce=gelişmişlik.ilce.str.strip()
gelişmişlik.skor=gelişmişlik.skor.str.strip()
gelişmişlik.kademe=gelişmişlik.kademe.str.strip()
gelişmişlik=gelişmişlik.set_index(gelişmişlik.ilce)
gelişmişlik.drop(['ilce','kademe'],axis=1,inplace=True)
Kmean = KMeans(n_clusters=40)
Kmean.fit(gelişmişlik)
gelişmişlik['skor_label']=Kmean.labels_
gelişmişlik['skor_label']=gelişmişlik.skor_label.astype('O')
gelişmişlik=gelişmişlik.reset_index()

data2=data2.merge(gelişmişlik[['ilce','skor','skor_label']],how='left',on='ilce').set_index(data.index)
data_sub=data2[data2.güncel_fiyat.isnull()]
data2.drop(['tarih',
            'Ilan_ID',
            'il',
#             'ilce',
            'semt',
#             'usd_fiyat',
#             'brütm',
#             'netm',
            'konum',
#               'bulunduğukat',
#              'binadakikatsayısı',
             'USD',
             'skor',
#              'odasayısı',
#             'skor_label',
#             'semt_label',
#             'yıl_ay'
            ],axis=1,inplace=True)
```

```python
#dummies dataframe oluşturuldu
data_dummies=pd.get_dummies(data2,prefix=None, prefix_sep='_', drop_first=False)
target=data2[data2.güncel_fiyat.notnull()].güncel_fiyat.values.copy()
data_dummies.drop(['güncel_fiyat'],axis=1,inplace=True)
data_dummies
```

# Parametre Tuning

```python
from sklearn.model_selection import train_test_split,KFold
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.preprocessing import StandardScaler
import optuna
import xgboost as xgb
sc=StandardScaler()
data_dummies=sc.fit_transform(data_dummies)
data_end=data_dummies[:len(target)]
# target_2=np.log1p(target)
x_train,x_valid,y_train,y_valid=train_test_split(data_end,target,test_size=0.1,random_state=4)


def objective(trial,data=data_end,target=target):
    
    train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.1,random_state=4)
    param = {
        'tree_method':'gpu_hist',  # this parameter means using the GPU when training our model to speedup the training process
        'lambda': trial.suggest_loguniform('lambda', 1e-3, 10.0),
        'alpha': trial.suggest_loguniform('alpha', 1e-3, 10.0),
        'colsample_bytree': trial.suggest_categorical('colsample_bytree', [0.3,0.4,0.5,0.6,0.7,0.8,0.9, 1.0]),
        'subsample': trial.suggest_categorical('subsample', [0.4,0.5,0.6,0.7,0.8,1.0]),
        'learning_rate': trial.suggest_categorical('learning_rate', [0.008,0.009,0.01,0.012,0.014,0.016,0.018, 0.02]),
        'n_estimators': 1000,
        'max_depth': trial.suggest_categorical('max_depth', [5,7,9,11,13,15,17,20]),
        'random_state': trial.suggest_categorical('random_state', [24, 48,2020]),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 300),
    }
    model = xgb.XGBRegressor(**param)  
    
    model.fit(train_x,train_y,eval_set=[(test_x,test_y)],early_stopping_rounds=100,verbose=False)
    
    preds = model.predict(test_x)
    
    rmse = mean_squared_error(test_y, preds,squared=False)
    
    return rmse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

## XGBOOST MODEL

```python
import torch
best_params=study.best_trial.params
best_params.update({
                    'n_estimators': 2500,
                    'max_depth': 15,
                     'min_child_weight': 1,})

if torch.cuda.is_available()==True :
    best_params.update({'tree_method':'gpu_hist'})
best_params
```

```python

params= {'lambda': 0.17577008330073987, 
         'alpha': 0.0018481875717251614,
         'colsample_bytree': 0.7, 
         'subsample': 0.8,
         'learning_rate': 0.02, 
         'max_depth': 15, 
         'random_state': 2,
         'min_child_weight': 1,
         'n_estimators':1000,
         'tree_method':'gpu_hist'
         }

import xgboost  as xgb
xgb_tuned=xgb.XGBRegressor(
    **best_params
#      **params
                          ).fit(x_train,y_train,verbose=True) 
#model train
print('Model train')
print('rmse: ',np.sqrt(mean_squared_error(y_train,xgb_tuned.predict(x_train))))
print('r2: ',r2_score(y_train,xgb_tuned.predict(x_train)))


print('Model test')
print('rmse :',np.sqrt(mean_squared_error(y_valid,xgb_tuned.predict(x_valid))))
print('r2 :',r2_score(y_valid,xgb_tuned.predict(x_valid)))
print('-----------------------------------------------------------------------')
submission_dum=data_dummies[-len(data[data.güncel_fiyat.isnull()]):]
# subb_m=sc.transform(subb_dum)
xgb_pred=xgb_tuned.predict(submission_dum)
print(xgb_pred.min())
print(xgb_pred.max())
```
